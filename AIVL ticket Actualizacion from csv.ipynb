```python
#Script for adding the component number to the assembly which contains it. For the moment just useful for Mechanical items (M-XXXX)
 
import os
from jira import JIRA
import csv
import pandas as pd

user = os.environ['JiraUser'] 
apikey = os.environ['JiraToken2']
server = 'https://iceyedev.atlassian.net/'

options = {
 'server': server
}

jira = JIRA(options, basic_auth=(user,apikey) )

df = pd.read_csv ('NewAIVLAssemblies18.csv', sep=';')
# print(df)
hp = pd.read_csv ('18read.csv', sep=';')

satellite="X18"
parts=[]
for index, A in df.iterrows():
    for P in range(1,A['amount']+1):
        if A['amount']==0:
            continue
        elif A['amount']==1:
            DescString=f"{A['Name']} for {satellite}"
        else:                
            DescString=f"{A['Name']} {P} for {satellite}"

        parts=A['mechparts'].split("*")


        for i in range(0,len(parts)):
            for index, B in hp.iterrows():
                if B['Name'] == parts[i]:
                    parts[i] = str(parts[i])+": "+str(B['componentID'])

        for x in parts:
            DescString+=f"\n\n{x}"
            
        print(f"{DescString}")
        
```
