```python
# -*- coding: utf-8 -*-
"""


@author: atiainen

Read the docs will guide you.
https://jira.readthedocs.io/en/master/

This script will make assemblies for satellites according to the CSV

"""


import os
from jira import JIRA
import csv
import pandas as pd
#Do not put your credentials here. At the moment use env variables, but when we have the secret vault, use it and please purge old tokens.
user = os.environ['JiraUser'] 
apikey = os.environ['JiraToken2']
server = 'https://iceyedev.atlassian.net/'

#Note, when using user specific token, the same token is for all your instances. Not just test or production. Act accordingly.
options = {
 'server': server
}

jira = JIRA(options, basic_auth=(user,apikey) )

df = pd.read_csv ('NewAIVLAssemblies18.csv', sep=';')
print(df)

satellite=""
parts=[]
#Just making one as then assigning to the corresponding epic is easy using bulk edit. Adding the the line for assigning to the correct epic directly will make that step unnecessary.
for i in range(18,19):
    
    satellite="X"+str(i)
    for index, A in df.iterrows():
        for P in range(1,A['amount']+1):
            if A['amount']==0:
            
                continue
            
            elif A['amount']==1:
                ticketSummary=f"{A['Summary']}"
                DescString=f"{A['Name']} for {satellite}"
            else:                
                ticketSummary=f"{A['Summary']} {P}"
                DescString=f"{A['Name']} {P} for {satellite}"
            
            parts=A['mechparts'].split("*")
            for x in parts:
                DescString+=f"\n\n{x}:"
            print(ticketSummary)
            print(f"{DescString}")
            print(A['componentID'])
            #issues aka tickets are simply objects with a lot of parameters called fields. See the documentation how to get/set those. Also the field naming can be very cryptic. 

        #Making new issue -- Uncomment this section to make changes in the production system


            new_issue = jira.create_issue(project='AIVL', summary=ticketSummary,
                               description=DescString, issuetype={'name': 'Story'})



            existingComponents = []

            existingComponents.append({'id': str(A['componentID'])})

            print(existingComponents)
            new_issue.update(fields={"components": existingComponents})
```
